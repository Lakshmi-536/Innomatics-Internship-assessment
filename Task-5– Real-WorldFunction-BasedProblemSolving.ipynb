```python
# 1: Social Media – Post Engagement Analyzer
def likes():
    t=0
    n=int(input("Enter number of days you want to consider for calculating total likes:"))
    for i in range(n):
        l=int(input(f"Enter day {i+1} likes:"))
        t+=l
    if t>=1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes()    
```

```python
# 2: Healthcare – Medicine Stock Alert System
def healthcare():
    stock=int(input("Enter current medicine stock level:"))
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")
healthcare()
```

```python
# 3: Agriculture – Rainfall Adequacy Checker
def agriculture():
    a=int(input("No.of rainfall readings: "))
    required_level=int(input("Enter required level: "))
    total=0
    for _ in range(a):
        rain=int(input("Enter rainfall reading: "))
        total+=rain
    if(total/a)>=required_level:
        print("Adequate Rainfall")
    else:
        print("Inadequate Rainfall")
agriculture()
```

```python
# 4: Social Media – Duplicate Account Detection
def media():
    names=["Lakshmi","Sri","Shiva","Parvati","Arjun","Priya","shiva"]
    t=len(names)
    names1=set(names)
    l=len(names1)
    if(t==l):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
media()
```

```python
# 6: Agriculture – Premium Crop Price Filter
def crop(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:", premium)
crop([1500, 2500, 1800, 3200, 2000])
```

```python
# 7: System Monitoring – Application Health Checker

def check_health(errors):
    print("Error Count:", errors)
    
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

# Example 
check_health(7)
```

```python
# 8: Banking – Daily Transaction Limit Checker
def check_transaction():
    amount=int(input("Enter anount:"))
    limit = 50000
    print("Transaction Amount:", amount)
    
    if amount <= limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")
check_transaction()
```

```python
# 9: E-Learning – Student Attendance Eligibility System

def check_attendance():
    n = int(input("Enter total classes: "))
    present = 0
    for i in range(n):
        if input("Enter P/A: ").upper() == "P":
            present += 1
    percentage = (present / n) * 100
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", "Eligible" if percentage >= 75 else "Not Eligible")

check_attendance()
```

```python
# 10: Smart Electricity Bill Analyzer
def calculate_bill(units):
    bill = 0 
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status
    
units = int(input("Enter units consumed: "))
bill, usage = calculate_bill(units)

print("Total Bill:", bill)
print("Usage Level:", usage)
```
