```python
# 1. User Login Check
username = input("Enter name: ")
password = int(input("Enter Password: "))
if(username=="admin" and password == 1234):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
passed=0
failed=0
for m in marks:
    if(m>=50):
        passed+=1
    else:
        failed+=1
print("No.of students passed:",passed,"No.of students failed:",failed)
```

```python
# 3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
new_array=[]
for name in names:
    new_array.append(name.strip().lower())
print(new_array)
```

```python
# 4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    print("length of message:",len(msg))
    if(len(msg)>10):
        flag=True
    else:
        flag=False
    print("flag:",flag)
```

```python
#  Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
cnt=0
for i in logs:
    if i=="ERROR":
        cnt+=1
print("Total error count :",cnt)
```
