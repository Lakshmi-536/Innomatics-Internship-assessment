```python
# 2: Inventory Reorder Alert System
def inventory(stock):
    for item,val in stock.items():
        if val<15:
            print(item,": Reorder Alert")
        else:
            print(item,": Stock OK")
stock = {
    "Laptop": 30,
    "Mouse": 25,
    "Keyboard": 15,
    "Monitor": 8,
    "Headphones": 10
}
inventory(stock)
```

```python
# 3: Student Result Processing System
def marks():
    marklist=list(map(int,input("Enter marks:").split(',')))
    total=0
    for i in marklist:
        total+=i
    avg=total/len(marklist)
    if avg>=50:
        print("Pass")
    else:
        print("Fail")
marks()
             
```

```python
# 4: Cab Fare Estimator with Retry Option
def farecalculation():
    dist=int(input("Enter distance in km:"))
    peak=input("Enter yes or no for peak hour:")
    amt=50+(dist*12)
    if peak.lower()=="yes":
        amt=amt+(0.25*amt)
    print("Cab Fare:",amt)
    
while True:
    farecalculation()
    retry=input("Retry: Yes/No")
    if retry.lower()=="no":
        break
```

```python
# 5: Employee Attendance Eligibility
def attendance():
    # P-Present , A-Absent
    attend=input("Enter P or A").split() 
    total=0
    for i in attend:
        if i=="P":
            total+=1
    if total/len(attend)>=0.75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance()
```

```python
# 6: Password Strength Checker
def password(string):
    s="@#$"
    d="0123456789"
    has_special=False
    has_digit=False
    if len(string)>=8:
        for char in string:
            if char in s:
                has_special=True
            if char in d:
                has_digit=True
        if has_special and has_digit:
            print("Strong")
    else:
        print("Weak")
        
string=input("Enter password:")
password(string)
```
