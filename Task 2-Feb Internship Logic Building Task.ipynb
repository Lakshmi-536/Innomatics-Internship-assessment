```python
# 1: Unique Words in a Sentence
st= "python is easy and python is powerful"
Set=set(st.split(' '))
print(Set)
print("length:",len(Set))
```

```python
# 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Start with a very small number for max_val (so any real value is bigger)
max_val = float('-inf')
max_key= ""
for key,val in employees.items():
    if val>max_val:
        max_val=val
        max_key=key
print("Highest Salary:",max_key,max_val)
```

```python
# 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
print(numbers)
maxi=numbers[0]
mini=numbers[0]
for num in numbers :
    if num>maxi:
        maxi=num
    elif num<mini:
        mini=num
print("Max:",maxi,"Min:",mini)
```

```python
#  4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
cnt=0
for i in prices:
    if i>1000:
        cnt+=1
print("Products above 1000:",cnt)
```

```python
5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
p_cnt=0
t=0
for a in attendance:
    if a=='P':
        p_cnt+=1
    t+=1
print("Attendance Percentage:",(p_cnt/t)*100)
```

```python
# 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
Unique_phone_numbers=set(phone_numbers)
print("Unique phone numbers:",Unique_phone_numbers)
```

```python
# 7: Count Character Frequency
string="Welcome To Python Programming"
dictionary={}
for char in string:
    if char in dictionary:
        dictionary[char]+=1
    else:
        dictionary[char]=1
print("dictionary is:",dictionary)
```
